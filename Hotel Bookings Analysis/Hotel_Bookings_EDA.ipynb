## Cancellation Analysis

```python
#cancelation rate
canceled_rate = len(df[df['is_canceled'] == 1]['is_canceled']) / len(df['is_canceled'])
print("The Cancelation Rate is {:.2f}".format(canceled_rate * 100))
```

```python
#cancelation by market segment
cancelation_by_market_segment = df.groupby('market_segment')['is_canceled'].mean()
plt.figure(figsize= (18, 6))
cancelation_by_market_segment.plot(kind = "bar", title = "cancelation by market segment", ylabel = "cancelation rate")
```

```python
#Cancellations by Year
import seaborn as sns
plt.figure(figsize= (18, 6))
sns.countplot(x='arrival_date_year', hue='is_canceled', data=df)
plt.title('Cancellations by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend(title='Is Canceled', loc='upper right')
plt.show()
```

```python
#Special Requests vs. Cancellations
plt.figure(figsize= (18, 6))
sns.boxplot(x='is_canceled', y='total_of_special_requests', data=df)
plt.title('Special Requests vs. Cancellations')
plt.xlabel('Is Canceled')
plt.ylabel('Total of Special Requests')
plt.show()
```

```python
#Cancellation by Country
cancellation_by_country = df.groupby('country')['is_canceled'].mean()
cancellation_by_country[:5]
```

```python
cancellation_by_country = df.groupby('country')['is_canceled'].mean()
cancellation_by_country.sort_values(ascending = False)[:5]
```

```python
#Repeated Guests and Cancellations
import matplotlib.pyplot as plt
repeated_guest_and_cancellations = df.groupby('is_repeated_guest')['is_canceled'].mean()
plt.figure(figsize= (18, 6))
repeated_guest_and_cancellations.plot(kind = "bar")
plt.title("Repeated Guests and Cancellations")
plt.xticks(ticks = [0, 1], labels = ["no", "yes"])
plt.xlabel('Is Repeated Guest')
plt.ylabel('Cancellation Rate')
plt.show()
```

##  Guest Demographics and Characteristics

```python
#Heatmap of Correlation Between Numerical Variables
plt.figure(figsize=(14, 10))
corr = df.select_dtypes(include = 'number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#the average number of adults, children, and babies per booking
avg_adults = df['adults'].mean()
avg_children = df['children'].mean()
avg_babies = df['babies'].mean()

print(f"Average Number of Adults: {avg_adults:.2f}")
print(f"Average Number of Children: {avg_children:.2f}")
print(f"Average Number of Babies: {avg_babies:.2f}")
```

```python
#classify travelers as 'family' or 'solo'
def selection(cols):
    if cols['babies'] > 0 or  cols['children'] > 0:
        return "Family"
    else:
        return "Solo"
df['traveler_type'] = df.apply(selection, axis = 1)
traveler_counts = df['traveler_type'].value_counts()
plt.figure(figsize= (18, 6))
traveler_counts.plot(kind='bar')
plt.title('Family vs Solo Travelers')
plt.xlabel('Traveler Type')
plt.ylabel('Number of Bookings')
plt.show()
```

```python
#meal classification
meal_classified = df['meal'].value_counts()
plt.figure(figsize= (18, 6))
meal_classified.plot(kind = "bar")
plt.title('meal classification')
plt.xlabel('meal type')
plt.ylabel('number of bookings')
plt.show()
```

```python
#cancellation by mean
cancellation_by_meal = df.groupby("meal")['is_canceled'].mean()
plt.figure(figsize= (18, 6))
cancellation_by_meal.plot(kind = "bar")
plt.title('cancellation by meals')
plt.xlabel('meal type')
plt.ylabel('number of cancellations')
plt.show()
```

<b>The guest demographics and characteristics analysis reveals the following key insights:

- **Average Number of Adults:** 1.86
- **Average Number of Children:** 0.10
- **Average Number of Babies:** 0.01

From this, it is clear that the majority of guests are adults, with very few children or babies traveling. 

Additionally, there is a higher presence of **solo travelers**, who prefer **Bed and Breakfast (BB)** as their meal option. The cancellation behavior does appear to be significantly influenced by **Food and Beverage (FB)** preferences, indicating that cancellations are not that strongly linked to meal choices.</b>

<b>The revenue analysis shows that Room Type A generates the highest revenue, totaling 24,617,818.19. The countries with the highest ADR are AIA (265.00), AND (202.65), and ABW (128.34). ## Customer Loyality

```python
#repeated guest and cancellations
repeated_guest = df[df['is_repeated_guest'] == 1]
print(f"Cancellation rate by repeated guest {repeated_guest['is_canceled'].mean() * 100:.2f}")
repeated_guest = df[df['is_repeated_guest'] == 0]
print(f"Cancellation rate by non repeated guest {repeated_guest['is_canceled'].mean() * 100:.2f}")
```

```python
guests_with_previous_cancellations = df[df['previous_cancellations'] > 0]
guests_without_previous_cancellations = df[df['previous_cancellations'] == 0]
prev_cancel_rate = guests_with_previous_cancellations['is_canceled'].mean()
print(f"Cancellation rate for guests with previous cancellations: {prev_cancel_rate:.2f}")
no_prev_cancel_rate = guests_without_previous_cancellations['is_canceled'].mean()
print(f"Cancellation rate for guests without previous cancellations: {no_prev_cancel_rate:.2f}")
```

<b>The cancellation analysis reveals that **repeated guests** have a lower cancellation rate of **14.49%**, compared to **non-repeated guests**, who have a significantly higher cancellation rate of **37.79%**. Furthermore, guests with **previous cancellations** show a cancellation rate of **0.92%**, while those without previous cancellations have a much lower rate of **0.34%**. This suggests that previous behavior and loyalty have a notable impact on cancellation trends.</b>

## Market Segment Performance

```python
#calculate cancellation rate and ADR
market_segment_stats = df.groupby('market_segment').agg(
    cancellation_rate=('is_canceled', 'mean'),
    avg_adr=('adr', 'mean')
)
print(market_segment_stats)
```

```python
#cancellation rates by market segment
plt.figure(figsize= (18, 6))
market_segment_stats['cancellation_rate'].plot(kind='bar')
plt.title('Cancellation Rate by Market Segment')
plt.xlabel('Market Segment')
plt.ylabel('Cancellation Rate')
plt.show()
```

```python
#ADR by market segment
plt.figure(figsize= (18, 6))
market_segment_stats['avg_adr'].plot(kind='bar')
plt.title('ADR by Market Segment')
plt.xlabel('Market Segment')
plt.ylabel('Average Daily Rate (ADR)')
plt.show()
```

```python
#ADR Distribution
plt.figure(figsize= (18, 6))
sns.histplot(df['adr'], bins=10, kde=True)
plt.title('ADR Distribution')
plt.xlabel('ADR')
plt.ylabel('Frequency')
plt.show()
```

<b>The market segment performance analysis reveals significant variations in both cancellation rate and average daily rate (ADR) across different segments:

Undefined Market Segment has the highest cancellation rate at 1.00, indicating a very high level of cancellations in this group. This segment also has the lowest ADR at 15.00.
Offline TA/TO has the highest ADR at 87.35, despite having a relatively high cancellation rate of 0.34.
Other segments show varied performance, with Direct having a relatively low cancellation rate of 0.15 and a higher ADR of 115.45, while Groups have the highest cancellation rate of 0.61 but an ADR of 79.48.

In summary, Undefined market segments are associated with the highest cancellations, while Offline TA/TO stands out for its higher ADR.</b>
