<h3><b> Make a table of all the producers and directors of each movie. Find the top 3 producers who have produced movies with the highest average RoI? </b></h3>

```python
columns = ['id', 'title', 'budget', 'revenue', 'crew']
data = df.loc[:, columns]
data[:5]
```

```python
import ast
def transform(item):
    if isinstance(item, str):
        try:
            return ast.literal_eval(item)
        except ValueError:
            return item
    return item 
data['crew'] = data['crew'].apply(transform)
```

```python
producers = []
directors = []
for rows in data['crew']:
    producer_names = []
    director_names = []
    if isinstance(rows, list):
        for names in rows:
            if names['job'] == 'Producer':
                producer_names.append(names['name'])
            elif names['job'] == 'Director':
                director_names.append(names['name'])
        producers.append(', '.join(producer_names))
        directors.append(', '.join(director_names))
    else:
        producers.append(np.nan)
        directors.append(np.nan)
data['Producers'] = producers
data['Directors'] = directors
```

```python
data['Producers'].fillna("No Producers", inplace=True)
data['Directors'].fillna("No Directors", inplace=True)
```

```python
data['Producers'] = data['Producers'].replace("", "No Producers")
data['Directors'] = data['Directors'].replace("", "No Directors")
```

```python
data_with_roi = data.drop(columns=['crew'])
```

```python
data_with_roi['roi'] = ((data_with_roi['revenue'] - data_with_roi['budget']) / data_with_roi['budget']) * 100
```

```python
data_with_roi.loc[:, 'roi'] = data_with_roi.loc[:, 'roi'].astype(int)
```

```python
avg_roi_by_pandd = data_with_roi.groupby(['title', 'Producers', 'Directors'])['roi'].mean().reset_index().sort_values(by='roi', ascending=False)
avg_roi_by_pandd
```

```python
top_3_pandd = avg_roi_by_pandd.sort_values(by='roi', ascending=False).head(3)
top_3_pandd
```

<h3> <b>Which actor has acted in the most number of movies? Deep dive into the movies, genres and profits corresponding to this actor. </b></h3>

```python
columns = ["id", "cast", "genres", "revenue", "budget", "original_title", "release_date"]
data = df.loc[:, columns]
data[:3]
```

```python
data = data.dropna(subset=['cast'])
```

```python
import ast
def transform(item):
    if isinstance(item, str):
        try:
            return ast.literal_eval(item)
        except ValueError:
            return item
    return item 
data['cast'] = data['cast'].apply(transform)
```

```python
actors = []  
for i in data.iloc[:, 1]:
    actor = None  
    if isinstance(i, list):  
        for j in i:
            if isinstance(j, dict):  
                if 'name' in j:  
                    actor = j['name']  
                    break  
    actors.append(actor)
data['actor'] = actors
```

```python
data = data.drop(columns = ['cast'])
```

```python
data['profit'] = data['revenue'] - data['budget']
actor_group = data.groupby('actor').agg(
    movie_count=('actor', 'size'),
    genres=('genres', 'unique'),
    revenue=('revenue', 'sum'),
    budget=('budget', 'sum'),
    original_title=('original_title', 'unique'),
    profit=('profit', 'sum')
).reset_index()
actor_group_sorted = actor_group.sort_values(by="movie_count", ascending=False)
actor_group_sorted[:3]
```

<h3><b>Top 3 directors prefer which actors the most? </b></h3>

```python
columns = ["id", "cast", "crew", "original_title"]
data = df.loc[:, columns]
data[:3]
```

```python
import ast
columns_to_be_transformed = ['cast', 'crew']
def transform(item):
    if isinstance(item, str):
        try:
            return ast.literal_eval(item)
        except ValueError:
            return item
    return item 
for i in columns_to_be_transformed:
    data[i] = data[i].apply(transform)
```

```python
actors = []  
for i in data.iloc[:, 1]:
    actor = None  
    if isinstance(i, list):  
        for j in i:
            if isinstance(j, dict):  
                if 'name' in j:  
                    actor = j['name']  
                    break  
    actors.append(actor)
data['actor'] = actors
data[:3]
```

```python
directors = [] 
for i in data.iloc[:, 2]:  
    director_name = None 
    if isinstance(i, list):
        for j in i:
            if isinstance(j, dict):  
                if j.get('job') == "Director":  
                    director_name = j.get('name')  
                    break 
    directors.append(director_name)  
data['director'] = directors 
data[:3]
```

```python
data_clean = data[data['actor'].notnull() & data['director'].notnull()]
```

```python
data_clean['actor_director'] = list(zip(data_clean['actor'], data_clean['director']))
data_clean[:5]
```

```python
actor_director_collab = data_clean.groupby('actor_director').size().reset_index(name='collab_count')
actor_director_collab
```

```python
actor_director_collab_sorted = actor_director_collab.sort_values(by='collab_count', ascending=False)
actor_director_collab_sorted.head(3)
```
